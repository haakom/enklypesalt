```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import math

# Define a custom colorblind friendly colormap
custom_cmap = [
    '#e69f00',   
    '#0072b2',
    '#56b4e9',
    '#cc79a7', 
    '#f0e442', 
    '#009e73', 
    '#000000'
]
plt.rcParams["axes.prop_cycle"] = cycler(color=custom_cmap)

# Seed random generator for reproducibility
rng = np.random.default_rng(42)

# Set n samples we want
n_samples = 256

# Generate data
weights_m = rng.normal(80, 10, n_samples//2)
weights_f = rng.normal(65, 10, n_samples//2)

heights_m = weights_m + rng.normal(100, 20, n_samples//2)
heights_f = weights_f + rng.normal(90, 10, n_samples//2)

#is_male = np.ones(n_samples, dtype=np.int8)
#is_male[n_samples//2:]-=1

# Create dataframe from data
df_w2h = pd.DataFrame({"height": np.concatenate([heights_m, heights_f]), 
                       "weight": np.concatenate([weights_m, weights_f])})
df_w2h = df_w2h[~df_w2h['weight'].between(75, 80, inclusive='left')]
new_row = pd.DataFrame({'height': [130, 220], 'weight': [75.1, 79.9]})
df_w2h = pd.concat([df_w2h, new_row], ignore_index=True)
print(df_w2h.head())
```

```python
plt.figure()
plt.scatter(df_w2h.weight, df_w2h.height)
plt.xlabel("weight (kg)")
plt.ylabel("height (cm)")
plt.savefig("figs/weight_height_dataset.png")
plt.show()
```

```python
def euclidian_distance(w1: float = None, w2: float = None):
    """
    Calculates the euclidian distance between two weight measurements w1 and w2

    returns: the calculated distance
    """
    return math.sqrt((w1-w2)**2)

def nearest_neighboor(dataset, new_weight: float= None):
    distances = []

    # Calculate distance between our new weight measurement and all other weight measurements we have in our data
    for idx, person in dataset.iterrows():
        # For each person in our dataset, we calculate the distance to the new weight and we store the height of this person
        distances.append(np.array([euclidian_distance(new_weight, person.weight), person.height]))

    # Sort the list of distances calculated
    distances.sort(key = lambda x: x[0])

    # Return the height of the person in our dataset that has the closest weight to the one we measured
    return distances[0][1]
```

```python
weights = np.arange(0, 180, 1) # Generate a range of new weights

# Use our functions to get the nearest weight in our dataset for every weight we just generated
predictions = [nearest_neighboor(df_w2h, wt) for wt in weights]

plt.figure()
plt.scatter(df_w2h.weight, df_w2h.height, label="person in dataset")
plt.plot(weights, predictions, color='red', label="model prediction")

plt.legend()

plt.xlabel("weight")
plt.ylabel("height")

plt.title("Nearest distance predictions vs. dataset")
plt.show()
plt.savefig("figs/nearest_neighboor.png")
```

```python
verification_weights = pd.DataFrame(df_w2h.sample(int(len(df_w2h)*0.1)))# Sample verification data

dataset_without_verification_weights = df_w2h.drop(verification_weights.index) # Remove the sample from the data used to generate predictions


# Use our functions to get the nearest weight in our dataset for every weight in our verification data
predictions = [nearest_neighboor(dataset_without_verification_weights, wt[1].weight) for wt in verification_weights.iterrows()]
```

```python
for idx, vp in enumerate(verification_weights.iterrows()):
    print(f'pred: {predictions[idx]:.2f}, true: {vp[1].height:.2f}, difference: {(predictions[idx]-vp[1].height):.2f}')
```

```python
# Calculate average difference
differences = []
for idx, vp in enumerate(verification_weights.iterrows()):
    differences.append(predictions[idx]-vp[1].height)

print(sum(differences)/len(differences))
```

```python
# Calculate root squared average distance
differences = []
for idx, vp in enumerate(verification_weights.iterrows()):
    differences.append((predictions[idx]-vp[1].height)**2)

print(math.sqrt(sum(differences)/len(differences)))
```
